```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load the data
data = pd.read_csv('../data/mushrooms.csv')

# first few rows of the dataset
data.head()

```

```python
# statistics
data.describe(include='all')
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# class distribution
sns.countplot(x='class', data=data)
plt.title('Class Distribution')
plt.show()
```

```python
# distribution of 'cap-shape'
plt.figure(figsize=(10, 4))
sns.countplot(x='cap-shape', hue='class', data=data)
plt.title('Cap Shape Distribution')
plt.show()
```

```python
# distribution of 'cap-surface'
plt.figure(figsize=(10, 4))
sns.countplot(x='cap-surface', hue='class', data=data)
plt.title('Cap Surface Distribution')
plt.show()
```

```python
#distribution of 'cap-color'
plt.figure(figsize=(10, 4))
sns.countplot(x='cap-color', hue='class', data=data)
plt.title('Cap Color Distribution')
plt.show()
```

```python
# distribution of 'bruises'
plt.figure(figsize=(10, 4))
sns.countplot(x='bruises', hue='class', data=data)
plt.title('Bruises Distribution')
plt.show()
```

```python
# distribution of 'odor'
plt.figure(figsize=(10, 4))
sns.countplot(x='odor', hue='class', data=data)
plt.title('Odor Distribution')
plt.show()
```
